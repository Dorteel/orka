# Test Agent and Sensor Management

Brief test of the agents.py module

```python
import sys
sys.path.insert(0, '/home/user/pel_ws/src/orvis/orka')

from rdflib import Graph
from graph_manager.agents import load_orka, add_agent
```

## Create Graph and Load ORKA

```python
# Create new graph
g = Graph()

# Load ORKA ontology
load_orka(g)

print(f"Graph size: {len(g)} triples")
```

## Add Robot with Sensors

```python
# Add robot with sensors
robot_sensors = ['temperature_sensor', 'distance_sensor', 'camera']
robot_uri = add_agent(g, 'pr2', 'robot', sensors=robot_sensors)

print(f"Created robot: {robot_uri}")
print(f"Graph size: {len(g)} triples")
```

## Add Human

```python
# Add human (no sensors)
human_uri = add_agent(g, 'operator1', 'human')

print(f"Created human: {human_uri}")
print(f"Graph size: {len(g)} triples")
```

## View Graph Content

```python
# Print turtle representation (first 30 triples)
triples = list(g.triples((None, None, None)))
print(f"Total triples: {len(triples)}")
print("\nFirst 20 triples:")
for i, (s, p, o) in enumerate(triples[:20]):
    print(f"{i+1}. {s.split('#')[-1]} -- {p.split('#')[-1]} --> {o.split('#')[-1] if isinstance(o, str) else o}")
```

```python
# Save to file
output_path = '/home/user/pel_ws/src/orvis/orka/output/agents_test.ttl'
g.serialize(destination=output_path, format='ttl')
print(f"Saved to: {output_path}")
```
